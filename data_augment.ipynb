```python
import os
from PIL import Image
import matplotlib.pyplot as plt
```

```python
def display_downsampled_image(image_path, downsampling_factor):
    # Open the original image
    with Image.open(image_path) as img:
        # Get the original dimensions
        original_width, original_height = img.size
        
        # Calculate the downsampled dimensions
        new_width = original_width // downsampling_factor
        new_height = original_height // downsampling_factor
        
        # Downsample the image
        downsampled_img = img.resize((new_width, new_height))
        
        # Display the original and downsampled images side by side using matplotlib
        fig, axes = plt.subplots(1, 2, figsize=(10, 5))
        
        # Original image
        axes[0].imshow(img)
        axes[0].set_title(f'Original ({original_width}x{original_height})')
        axes[0].axis('off')
        
        # Downsampled image
        axes[1].imshow(downsampled_img)
        axes[1].set_title(f'Downsampled ({new_width}x{new_height})')
        axes[1].axis('off')
        
        # Show the plot
        plt.show()
```

```python
sample_path = '10Knots/Bowline Knot/IMG_6469.jpg'
```

```python
display_downsampled_image(sample_path, 8)
```
