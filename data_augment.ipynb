```python
import os
from PIL import Image, ImageFilter
import matplotlib.pyplot as plt
```

```python
def get_size(image):
    return image.size
```

```python
def get_downsampled_image(image_path, downsample_factor):
    image = Image.open(image_path)
    return image.resize((image.width // downsample_factor, image.height // downsample_factor))
```

```python
def rotate_image(image, angle):
    return image.rotate(angle)
```

```python
def flip_image(image):
    return image.transpose(Image.FLIP_LEFT_RIGHT)
```

```python
def color_space_conversion(image, mode):
    # choices of mode: 'L', 'RGB', 'RGBA', 'CMYK', 'YCbCr', 'I', 'F'
    return image.convert(mode)
```

```python
def kernel_convolution(image, kernel):
    return image.filter(kernel)
```

```python
def display_image(image):
    plt.imshow(image)
    plt.show()
```

```python
sample_path = '10Knots/Bowline Knot/IMG_6469.jpg'
```

```python
sample_image = Image.open(sample_path)
get_size(sample_image)
```

```python
get_size(get_downsampled_image(sample_path, 8))
```

```python
get_size(kernel_convolution(get_downsampled_image(sample_path, 8), ImageFilter.CONTOUR))
```

```python
display(kernel_convolution(get_downsampled_image(sample_path, 8), ImageFilter.SMOOTH_MORE))
```
