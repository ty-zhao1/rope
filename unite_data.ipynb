```python
import os
import sys
from PIL import Image
```

```python
def convert_images_to_jpg(base_folder):
    """
    Convert all images in a folder (and subfolders) to JPG format.
    
    Args:
        base_folder (str): The path to the folder containing images.
    """
    for root, _, files in os.walk(base_folder):
        for file in files:
            if file == ".DS_Store":
                continue
            file_path = os.path.join(root, file)
            try:
                # Open the image
                with Image.open(file_path) as img:
                    # Ensure the image is in RGB mode (required for JPG)
                    img = img.convert("RGB")
                    
                    # Construct the new file name with a .jpg extension
                    new_file_path = os.path.splitext(file_path)[0] + ".jpg"
                    
                    # Save the image as JPG
                    img.save(new_file_path, "JPEG")
                    print(f"Converted: {file_path} -> {new_file_path}")
                    
                    # Optionally delete the original file
                    # os.remove(file_path)
            
            except Exception as e:
                # Handle errors (e.g., if the file is not an image)
                print(f"Error converting {file_path}: {e}")

# Replace this with your folder path
base_folder_path = "CS230_extra_data"
convert_images_to_jpg(base_folder_path)
```

```python
import os
import shutil

def copy_images_excluding_webp(source_folder):
    """
    Copy all images from the source folder to a new folder named '<source_folder>_cleaned',
    maintaining the folder structure and excluding '.webp' files.

    Args:
        source_folder (str): Path to the source folder containing images.
    """
    # Define the new directory
    parent_dir, source_name = os.path.split(source_folder.rstrip(os.sep))
    target_folder = os.path.join(parent_dir, f"{source_name}_cleaned")
    
    # Walk through the directory structure
    for root, dirs, files in os.walk(source_folder):
        # Compute the relative path from the source folder
        relative_path = os.path.relpath(root, source_folder)
        # Create the corresponding directory in the target folder
        target_path = os.path.join(target_folder, relative_path)
        os.makedirs(target_path, exist_ok=True)
        
        for file in files:
            file_ext = os.path.splitext(file)[-1].lower()
            
            # Exclude .webp files
            if file_ext == ".webp":
                continue
            
            # Full source and destination file paths
            source_file_path = os.path.join(root, file)
            target_file_path = os.path.join(target_path, file)
            
            try:
                # Copy the file
                shutil.copy2(source_file_path, target_file_path)
                print(f"Copied: {source_file_path} -> {target_file_path}")
            except Exception as e:
                print(f"Error copying {source_file_path}: {e}")

    print(f"All non-webp images copied to: {target_folder}")

# Replace this with your source folder path
source_folder_path = "CS230_extra_data"
copy_images_excluding_webp(source_folder_path)
```

```python
# compare the number of files in the original and cleaned folders
def count_files_in_folder(folder):
    """
    Count the number of files in a folder (including subfolders).
    
    Args:
        folder (str): The path to the folder.
    
    Returns:
        int: The number of files in the folder.
    """
    file_count = 0
    for root, _, files in os.walk(folder):
        file_count += len(files)
    return file_count

# Count the number of files in the original and cleaned folders
original_count = count_files_in_folder(source_folder_path)
cleaned_count = count_files_in_folder(f"{source_folder_path}_cleaned")

print(f"Original folder: {original_count} files")
print(f"Cleaned folder: {cleaned_count} files")
```

```python
import os
import shutil

def merge_directories(source1, source2, target):
    """
    Merge two directories into a new directory, maintaining the folder structure.

    Args:
        source1 (str): Path to the first source directory.
        source2 (str): Path to the second source directory.
        target (str): Path to the target directory where merged content will be stored.
    """
    def copy_files(source, target):
        for root, _, files in os.walk(source):
            # Compute the relative path from the source directory
            relative_path = os.path.relpath(root, source)
            target_path = os.path.join(target, relative_path)
            
            # Ensure the target directory exists
            os.makedirs(target_path, exist_ok=True)
            
            for file in files:
                source_file_path = os.path.join(root, file)
                target_file_path = os.path.join(target_path, file)

                if os.path.exists(target_file_path):
                    # Handle file name conflicts (overwrite, skip, or rename)
                    target_file_path = resolve_conflict(target_file_path)
                
                # Copy the file
                shutil.copy2(source_file_path, target_file_path)
                print(f"Copied: {source_file_path} -> {target_file_path}")

    def resolve_conflict(file_path):
        """
        Resolve file name conflicts by renaming the file with a unique suffix.
        
        Args:
            file_path (str): Path to the conflicting file.
        
        Returns:
            str: New file path with a unique suffix.
        """
        base, ext = os.path.splitext(file_path)
        counter = 1
        while os.path.exists(file_path):
            file_path = f"{base}_{counter}{ext}"
            counter += 1
        return file_path

    # Copy files from both source directories
    print("Copying from source1...")
    copy_files(source1, target)
    print("Copying from source2...")
    copy_files(source2, target)

    print(f"Merge complete. Merged content saved in: {target}")

# Replace these with your actual directory paths
source1_path = "10Knots"
source2_path = "CS230_extra_data_cleaned"
target_path = "10Knots_augmented"

merge_directories(source1_path, source2_path, target_path)
```

```python
def delete_ds_store_files(directory):
    """
    Recursively delete all .DS_Store files from the given directory and its subdirectories.

    Args:
        directory (str): Path to the directory to clean.
    """
    for root, _, files in os.walk(directory):
        for file in files:
            if file == ".DS_Store":
                file_path = os.path.join(root, file)
                try:
                    os.remove(file_path)
                    print(f"Deleted: {file_path}")
                except Exception as e:
                    print(f"Error deleting {file_path}: {e}")

# Replace this with your directory path
directory_path = "10Knots_augmented"
delete_ds_store_files(directory_path)
```
